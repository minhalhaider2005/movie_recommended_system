```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
)
```

```python
def convert(obj):
    L = [] 
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

# Convert Genres Column

```python
movies['genres'] = movies['genres'].apply(convert) 
```

```python
movies.head(2)
```

# Convert keywords Column

```python
movies['keywords'] = movies['keywords'].apply(convert) 
```

```python
movies.head(2)
```

# Convert Cast Column only 3 popular actor

```python
def convert3(obj):
    L = [] 
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter += 1
        else:
            break 
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(2)
```

# Only find the director name of the Crew Column

```python
def fetch_director(obj):
    L = [] 
    for i in ast.literal_eval(obj):
        if i ['job'] == 'Director':
            L.append(i["name"])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

# Split the overview column in to the list formate

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

# Merge the row data Sam Worthington to SamWorthington

```python
movies['cast'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x]) 
```

# Create a new column tag and concatenate four columns(cast,genres,crew,keywords)

# Import a Library

```python
from sklearn.feature_extraction.text import CountVectorizer
```

# Convert the tag into vector form

```python
cv = CountVectorizer(max_features = 5000,stop_words='english') 
```

```python
vector = cv.fit_transform(ne_df['tag']).toarray()
```

# Stemming apply

```python
from nltk.stem.porter import PorterStemmer
Ps =  PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(Ps.stem(i))
    return " ".join (y)
```

```python
ne_df['tag'] = ne_df['tag'].apply(stem)
```

# Calcualte the similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

 # Check the shape of the vector

```python
cosine_similarity(vector).shape
```

# Fetch the top 5 movies after recommended

```python
def recommend(movie):
    movie_index = ne_df[ne_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(ne_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
